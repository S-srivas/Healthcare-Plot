```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from pylab import bar, show, plot

df=pd.read_csv('Medicare.csv', parse_dates=['LocationAbbr'], index_col='LocationAbbr')
df1 = df.drop(['LocationDesc','DataSource','PriorityArea1','PriorityArea2','PriorityArea3','PriorityArea4','Category','Topic','Indicator','Data_Value_Type','Data_Value_Unit','Data_Value','Data_Value_Alt','Data_Value_Footnote_Symbol','Data_Value_Footnote','LowConfidenceLimit','HighConfidenceLimit','Break_Out_Category','Break_Out','CategoryId','TopicId','IndicatorID','Data_Value_TypeID','BreakOutCategoryId','BreakOutId','LocationID','GeoLocation'], axis = 1)
df_1 = pd.read_csv('Motor.csv', parse_dates=['Number of Vehicles Involved'], index_col='Year')
df2 = df_1.drop(['Crash Descriptor', 'Time', 'Date', 'Day of Week', 'Police Report','Lighting Conditions','Municipality','Collision Type Descriptor','County Name','Road Descriptor','Weather Conditions','Traffic Control Device','Road Surface Conditions','DOT Reference Marker Location','Pedestrian Bicyclist Action','Event Descriptor'], axis = 1)
#print (df1)
print (df2)

d3 = str(df1)
d4 = str(df2)
file1 = open('New.csv', 'w')
for line in df1:
    file1.write(d3)
file1.close()

file2 = open('Motor_New.csv', 'w')
for line in df2:
    file2.write(d4)
file2.close()
```

```python
def plot_plots(df, x, y, z, title="", xlabel='State', ylabel='Average Spending of Hospital', m1_tlabel='', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y, color='purple')
    plt.plot(x, z, color='red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    
    plt.show()
# Here, the two datdaframes are combined
#df_two = pd.concat([df, df_smallcols])
plot_plots(df, x=df.index, y=df.Year, z=df2, title='Average Spending in Hospitals Because of Claims in States')
```

```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression

data = pd.read_csv('New.csv')  # load data set
#df2 = data.drop(['Crash Descriptor', 'Time', 'Date', 'Day of Week', 'Police Report','Lighting Conditions','Municipality','Collision Type Descriptor','County Name','Road Descriptor','Weather Conditions','Traffic Control Device','Road Surface Conditions','DOT Reference Marker Location','Pedestrian Bicyclist Action','Event Descriptor'], axis = 1)
#d = str(df2)
#df = int(d)
e = data.fillna(0)
c = e.replace([np.inf, -np.inf], 0)
X = c.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = c.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```
